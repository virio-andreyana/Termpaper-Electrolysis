```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_res = pd.read_csv(f"../data/residualcapacity.csv")
df_life = pd.read_csv(f"../data/technologylifetime.csv")
```

```python
df_m = pd.DataFrame(columns=df_res.columns)

for t in df_res.technology.unique():
    for r in df_res.region.unique():
        A = df_res.query(f"year == 2020 & technology == @t & region == @r").value.sum()
        B = df_res.query(f"year == 2030 & technology == @t & region == @r").value.sum()
        C = df_res.query(f"year == 2040 & technology == @t & region == @r").value.sum()
        D = df_res.query(f"year == 2050 & technology == @t & region == @r").value.sum()
        if A == 0 and B == 0 and C == 0 and D == 0:
            continue
            
        l = df_life.query(f"technology == @t").value.item()
        
        if l == 15:
            df_m.loc[len(df_m)] = [2020,r,t,A]
            df_m.loc[len(df_m)] = [2030,r,t,B]
            df_m.loc[len(df_m)] = [2040,r,t,C]
            df_m.loc[len(df_m)] = [2050,r,t,D]
        
        elif l == 20:
            df_m.loc[len(df_m)] = [2010,r,t,A-B+C-D]
            df_m.loc[len(df_m)] = [2020,r,t,B-C+D]
            df_m.loc[len(df_m)] = [2030,r,t,C-D]
            df_m.loc[len(df_m)] = [2040,r,t,D]
            
        elif l == 30:
            df_m.loc[len(df_m)] = [2000,r,t,A-B+D]
            df_m.loc[len(df_m)] = [2010,r,t,B-C]
            df_m.loc[len(df_m)] = [2020,r,t,C-D]
            df_m.loc[len(df_m)] = [2030,r,t,D]
            
        elif l == 40:
            df_m.loc[len(df_m)] = [1990,r,t,A-B]
            df_m.loc[len(df_m)] = [2000,r,t,B-C]
            df_m.loc[len(df_m)] = [2010,r,t,C-D]
            df_m.loc[len(df_m)] = [2020,r,t,D]
        
df_m
```
